```python
# 파일 불러오기

import pandas as pd
df= pd.read_csv('water_accident_new.csv')
df.head()
```

```python
# NaN값 0으로 대체하기

df['time_unit_rainqty'] = df['time_unit_rainqty'].fillna(0)
df.head()
```

```python
# 연도별 사고수

import matplotlib.pyplot as plt

# 'dclr_yr' 열을 기준으로 그룹화하고 각 그룹의 개수를 계산
year_counts = df.groupby('dclr_yr')['dclr_yr'].count()

# 막대 그래프를 생성
plt.bar(year_counts.index, year_counts.values)

# 그래프 제목과 축 레이블을 설정
plt.title('Number of Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')

# 그래프 표시
plt.show()
```

```python
# 계절별 사고 수

# 'season_se_nm' 열을 기준으로 그룹화하고 각 그룹의 개수를 계산
season_counts = df.groupby('season_se_nm')['season_se_nm'].count()

# 막대 그래프를 생성
plt.bar(season_counts.index, season_counts.values)

# 그래프 제목과 축 레이블을 설정
plt.title('Number of Accidents by Season')
plt.xlabel('Season')
plt.ylabel('Number of Accidents')

# 그래프 표시
plt.show()
```
