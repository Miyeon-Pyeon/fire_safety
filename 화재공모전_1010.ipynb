```python
# 파일 불러오기

import pandas as pd
df= pd.read_csv('water_accident_new.csv')
df.head()
```

```python
df = df.drop(['sigungu_nm','time_unit_rainqty'], axis=1)
df
```

```python
# prcs_result_se_nm열 영어로 변경하기

# prcs_result_se_nm 열의 한글을 영어로 바꾸기 위한 딕셔너리 생성
translation_dict = {
    '인명구조': 'Rescue',
    '안전조치': 'Action',
    '인명검색': 'Search',
    '기타': 'Others'
}

# map 함수를 사용하여 prcs_result_se_nm 열의 값을 영어로 변환
df['prcs_result_se_nm'] = df['prcs_result_se_nm'].map(translation_dict)

df.head()
```

```python
# season_se_nm 영어로 변경하기

# season_se_nm 열의 한글을 영어로 바꾸기 위한 딕셔너리 생성
translation_dict_s = {
    '봄': 'Spring',
    '여름': 'Summer',
    '가을': 'Fall',
    '겨울': 'Winter'
}

# map 함수를 사용하여 season_se_nm 열의 값을 영어로 변환
df['season_se_nm'] = df['season_se_nm'].map(translation_dict_s)

df.head()
```

```python
# daywk 영어로 변경하기

# daywk 열의 한글을 영어로 바꾸기 위한 딕셔너리 생성
translation_dict_d = {
    '월요일': 'Mon',
    '화요일': 'Tue',
    '수요일': 'Wed',
    '목요일': 'Thu',
    '금요일': 'Fri',
    '토요일': 'Sat',
    '일요일': 'Sun'
}

# map 함수를 사용하여 daywk 열의 값을 영어로 변환
df['daywk'] = df['daywk'].map(translation_dict_d)

df.head()
```

```python
# 연도별 사고수

import matplotlib.pyplot as plt

# 'dclr_yr' 열을 기준으로 그룹화하고 각 그룹의 개수를 계산
year_counts = df.groupby('dclr_yr')['dclr_yr'].count()

# 막대 그래프를 생성
plt.bar(year_counts.index, year_counts.values)

# 그래프 제목과 축 레이블을 설정
plt.title('Number of Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')

# 그래프 표시
plt.show()
```

```python
# 계절별 사고 수

# 'season_se_nm' 열을 기준으로 그룹화하고 각 그룹의 개수를 계산
season_counts = df.groupby('season_se_nm')['season_se_nm'].count()


plt.rcParams['font.family'] ='Gulim'
plt.rcParams['axes.unicode_minus'] =False

# 막대 그래프를 생성
plt.bar(season_counts.index, season_counts.values)

# 그래프 제목과 축 레이블을 설정
plt.title('Number of Accidents by Season')
plt.xlabel('Season')
plt.ylabel('Number of Accidents')

# 그래프 표시
plt.show()
```
